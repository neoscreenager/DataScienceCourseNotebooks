```python
import matplotlib.pyplot as plt
from matplotlib import style
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
df_auto = pd.read_csv('auto_data.csv')
```

```python
df_auto.head()
```

```python
# replace origin value with corrosponding continent value
# 1-USA, 2-Europe, 3-Asia
def origin_code_to_name(code):
    if code == 1:
        return 'USA'
    elif code == 2:
        return 'Europe'
    elif code == 3:
        return 'Asia'
    else:
        return 'Unknown'

# use apply function to apply this funciton to all the values of origin column
df_auto['origin'] = df_auto['origin'].apply(origin_code_to_name)
```

```python
df_auto.head(30)
```

```python
# cerate a pair plot
sns.pairplot(df_auto[['mpg','weight','origin']], hue='origin', height=2)
```

```python
cause_of_death = ['Chronic Disease', 'Unintetional Injuries', 'Alzheimers', 'Infuenza and Pneumonia', 'Sepsis','Others']
percentile = [62, 5, 4, 2, 1 , 26]
```

```python
plt.figure(figsize=(5,5))
explode = (0.05, 0,0,0,0,0) # explode the largest percentile, so that it will be sliced out in the pie chart
plt.pie(percentile, labels=cause_of_death, explode=explode, autopct='%1.1f%%', startangle=70)
plt.axis('equal')
plt.title('Ohio state 2012 - Leadinng causes of Deaths')
```
