```python
import numpy as np

def is_inside_the_board(x , y):
    """
        is_inside_the_board checks if the coordinates of a possible position of the knight falls within the board.
    """
    if(x>=0 and x<=7 and y>=0 and y<=7):
        return True
    else:
        return False

def find_min_steps(initial_position, target_position, chess_board):
    """
        find_min_steps takes the initial and target coordinates of the knight and the 8x8 chess board as the third argument.
        The chess board is initialized with the flag 0, which indicates that the positions are all not visited. The function
        begins with marking the initial position of the knight as 1 i.e. visited, then follow these steps :
            1)Initialize the array of possible x and y steps that a knight can take on the board.
            2)Loop through the chess board for max 6 times, as for a 8x8 board, it takes max 6 moves to reach from any source position
              to any destination position.
            3)Within each loop, the x and y positions of the knight will be increased to make a move, then destination board position
              will be compared to the destination board position.
            4)The board position will be marked as visited (1), a move counter will be incremented by one
              and loop will continue until it reaches the destination board position.
            5)Once the destination board position is reached, the total number of move in this loop will be stored in a list.
            6)At the end of the while loop, the list of all the number of moves that the knight made to reach will be stored
              and the first element of this list is returned as the min number of steps.
    """
    # all possible movements for the knight
    dx = [2, 1, -1, -2, -2, -1, 1, 2];
    dy = [1, 2, 2, 1, -1, -2, -2, -1];
    
    moves = []
    for i in range(6):
        chess_board = np.zeros((8,8), dtype = int)
        # flag the initial position of the knight as visited
        chess_board[initial_position[0]][initial_position[1]] = 1
        temp_position = initial_position.copy()
        # iterate through all possible moves until target position is reached
        POSITIONS_EXPLORED = 0
        move_count = 0
        while(POSITIONS_EXPLORED < 63): 
            for k in range(8):
                X = temp_position[0] + dx[k]
                Y = temp_position[1] + dy[k]
                if(is_inside_the_board(X, Y) and chess_board[X][Y]==0):
                    temp_position[0] = X
                    temp_position[1] = Y
                    chess_board[temp_position[0]][temp_position[1]] = 1
                    print(X, Y)
                    move_count += 1
                    POSITIONS_EXPLORED += 1
                    if(temp_position[0]==target_position[0] and temp_position[1]==target_position[1]):
                        # since we have reached successfully to the target position, we will append the total number of moves
                        # we made to our list of moves
                        moves.append(move_count)
                        print(move_count)
                        print("moves array inside innermost for loop:", moves, POSITIONS_EXPLORED)
                        break
            
            else:
                continue
            break
                        
    moves.sort()
    return moves
    
def solution(src, dest):
    if(src>=0 and dest<=63):
        chess_board = np.zeros((8,8), dtype = int)
        i = 0
        for row in range(0,8):
            for column in range(0,8):
                chess_board[row][column] = i
                i += 1
        row_index_source, column_index_source = np.where(chess_board == src)
        row_index_dest, column_index_dest = np.where(chess_board == dest)
        knight_position = [int(row_index_source), int(column_index_source)]
        target_position = [int(row_index_dest), int(column_index_dest)]
        
        # create a position chess board for keeping track of positions of kinght's movement
        position_board = np.zeros((8,8), dtype = int)
        print("Sending {} as initial positon and {} as target position".format(knight_position, target_position))
        print(find_min_steps(knight_position, target_position, position_board))
    else:
        print("source and destinations should be within the board")

```
