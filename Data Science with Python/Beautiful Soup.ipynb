### Data parsing is the process of taking data in one format and transforming it to another format.

## Prob Statement-: Say we want to find the information about the countries and their capitals from say wikipedia,then as we do not have the free API's to access the information,we try to retrieve the information using the url of the website,and convert  the data from the html format using bs4

### A website's data is stored on a database, and then this data is presented on web pages. The data presentation includes database connectivity and database querying. The data is displayed over the internet in the form of web pages (HTML pages) 

```python
url = "https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_and_their_capitals_in_native_languages"
```

```python
#retreiving information and storing the information in variable res
res = requests.get(url)      #the get request to this url,gives us the html code
```

```python
#checking the status of retreival
res.status_code          #a status code of 200 means,okay,we will be getting the data 
```

```python
#Now we will retrieve the information from the website and store the content as text
text = res.content
```

```python
#We view the text variable and see an html output displayed--first line of output
text           #we can go to view page source(from the wikipedia page) to see this html output,as displayed below.
```

## Now we will use bs4 which will help create a parser for html


```python
#now we will use class beautiful soup from the module bs4 and pass the text in it which will create object soup
soup = bs4.BeautifulSoup(text)
```

```python
#lets check the object type of the variable 'soup'
type(soup)
```

```python
#To find the title of the html page
soup.title
```

```python
#to get the first/highest heading of the html page
soup.h1
#go to wiki page-inspect-we will get the title under head
```

## Till now we have found the information on countries from a wiki page by scraping through beautiful soup,now we want to store all this information in a csv file.We will first create an empty dataframe and then store all the information in a csv file. For that we import the library pandas

```python
import pandas as pd
```

```python
#we have an empty dataframe with only the columns,and we will populate the df with values later

df = pd.DataFrame({}, columns=['Country','Capital','Language'])
```

```python
df
```

## now to append values to the empty df,we need another df/list/dict(check attributes of append), so we will create a dict in our last looping code

```python
#Be careful with indentation post ":" in loops.A dict will store the o/p of the for loop


for tr in D:
    all_tds = tr.find_all('td')
     
    country = all_tds[0].text
    capital = all_tds[1].text
    language = all_tds[4].text
    
    dic = {'Country': country,'Capital': capital,'Language': language}
    #the key and values have the 1st letter in different case
    df = df.append(dic,ignore_index=True)
    
    print(country,"==",capital,"==",language)
    print()
```

```python
df
```

```python
df.shape
```

```python
#Last element of the 1st row under language column

df['Language'][0][-1]
```

## Now we do not want \n in language to reflect

```python
#So,we take everything excepting the last element
df['Language'][0][:-1]
```

## Now to remove the element \n from the languages,we use the lamda function

```python
df.head()
```

```python
df['Language'] =df['Language'].apply(lambda x:x[:-1])
```

```python
df
```

```python
#Index is false as we dont want the serial no to appear
df.to_csv("countries.csv",index=False)
```
