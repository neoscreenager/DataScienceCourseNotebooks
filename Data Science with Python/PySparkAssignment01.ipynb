```python
import findspark
findspark.init()
```

```python
import pyspark
```

```python
sc = pyspark.SparkContext()
```

```python
text_data_file = sc.textFile('Amazon_text_dataset.txt')
```

```python
from operator import add
```

```python
word_count_filter = text_data_file.filter(lambda x:len(x)>0)
```

```python
word_count_filter.count()
```

```python
#use flat map to split each line into words
word_flat_map = word_count_filter.flatMap(lambda x:x.split())
```

```python
#map each key with the value 1 using map function
word_mapper = word_flat_map.map(lambda x:(x,1))
```

```python
word_count = word_mapper.reduceByKey(add)
```

```python
word_count.first()
```

```python
#view first 50 key value pairs
type(word_count)
word_count.collect()
```
